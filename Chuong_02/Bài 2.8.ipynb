```python
import json

nhan_vien = {
    "ten": "Nguyễn Văn A",
    "tuoi": 25,
    "dia_chi": "01 Lê Duẩn, Hà Nội",
    "ky_nang": ["Python", "SQL", "XML"]
}

#Chuyển đối tượng Python sang chuỗi JSON
json_data = json.dumps(nhan_vien, ensure_ascii=False)

# In ra chuỗi JSON
print(json_data)

# In ra tất cả các giá trị trong đối tượng Python
print("\n Các giá trị trong đối tượng Python:")
for value in nhan_vien.values():
    print(value)
```
