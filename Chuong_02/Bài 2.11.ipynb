```python
import json
from datetime import datetime

# Danh sách giao dịch
giaodich_list = []

while True:
    loai = input("Nhập loại giao dịch (tiền/vàng): ")
    so_tien = input("Nhập số tiền: ")
    ghi_chu = input("Nhập ghi chú: ")

    # Tạo 1 giao dịch dưới dạng dict
    gd = {
        "loai": loai,
        "so_tien": so_tien,
        "ghi_chu": ghi_chu
    }

    # Thêm vào danh sách
    giaodich_list.append(gd)

    tiep = input("Bạn có muốn nhập thêm giao dịch không? (c/k): ")
    if tiep.lower() == 'k':
        break

# Hỏi người dùng có muốn ghi vào file JSON không
chon = input("Bạn có muốn ghi vào tập tin JSON không? (1: Có, 0: Không): ")

if chon == '1':
    # Lấy thời gian hiện tại
    now = datetime.now()
    ten_file = now.strftime("%Y-%m-%d-%H-%M-%S.json")

    # Ghi vào file JSON
    with open(ten_file, "w", encoding="utf-8") as f:
        json.dump(giaodich_list, f, ensure_ascii=False, indent=4)

    print(f"Đã ghi danh sách giao dịch vào file {ten_file}")
else:
    print("Không ghi vào tập tin.")
```
