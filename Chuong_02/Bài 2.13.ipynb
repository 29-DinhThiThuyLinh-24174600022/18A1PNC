```python
import requests

url = "https://jsonplaceholder.typicode.com/comments?postId=1"

response = requests.get(url)

data = response.json()

print("Tổng số bài post (comment) nổi bật:", len(data))
print("=" * 60)

print("Danh sách 3 bài post nổi bật đầu tiên:\n")
for post in data[:3]:
    print("Post ID:", post["postId"])
    print("ID:", post["id"])
    print("Tên người bình luận:", post["name"])
    print("Email:", post["email"])
    print("Nội dung:", post["body"])
    print("-" * 60)
```
