```python
import json

# Dữ liệu JSON 
data_json = '''
{
    "cong_ty": {
        "ten": "Công ty TNHH Đất Việt",
        "dia_chi": "abc Giải Phóng - Hà Nội"
    },
    "nhan_vien": [
        {"ten": "Nguyễn Văn A", "don_vi": "Đơn vị A1"},
        {"ten": "Trần Thị B", "don_vi": "Đơn vị A2"},
        {"ten": "Lê Văn C", "don_vi": "Đơn vị A1"},
        {"ten": "Phạm Thị D", "don_vi": "Đơn vị A3"},
        {"ten": "Vũ Văn E", "don_vi": "Đơn vị A4"},
        {"ten": "Ngô Thị F", "don_vi": "Đơn vị A2"},
        {"ten": "Đinh Văn G", "don_vi": "Đơn vị A4"}
    ]
}
'''

# Chuyển JSON sang đối tượng Python
data = json.loads(data_json)

#Lấy thông tin công ty và danh sách nhân viên
ten_cty = data["cong_ty"]["ten"]
dia_chi = data["cong_ty"]["dia_chi"]
ds_nv = data["nhan_vien"]
tong_nv = len(ds_nv)

# Thống kê nhân viên theo đơn vị
thong_ke = {}
for nv in ds_nv:
    don_vi = nv["don_vi"]
    thong_ke[don_vi] = thong_ke.get(don_vi, 0) + 1
print("Kết quả thực hiện chương trình:")
print(f"Tên công ty: {ten_cty}")
print(f"Địa chỉ: {dia_chi}")
print("----- Thống kê nhân viên theo đơn vị -----")

for i, (don_vi, so_nv) in enumerate(thong_ke.items(), start=1):
    ty_le = round(so_nv / tong_nv * 100, 2)
    print(f"{i}./Tên đơn vị: {don_vi}")
    print(f"   - Số nhân viên: {so_nv}")
    print(f"   - Tỷ lệ: {ty_le}%\n")
```
