```python
import threading

# Hàm mà mỗi luồng sẽ chạy
def in_ten_luong():
    print("Tên luồng:", threading.current_thread().name)

# Tạo danh sách để lưu các luồng
threads = []

# Tạo 5 luồng
for i in range(5):
    t = threading.Thread(target=in_ten_luong, name=f"Luong-{i+1}")
    threads.append(t)
    t.start()

# Chờ các luồng chạy xong
for t in threads:
    t.join()
```
