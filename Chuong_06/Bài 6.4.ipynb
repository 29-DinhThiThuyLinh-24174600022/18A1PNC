```python
import threading

# Biến toàn cục để lưu kết quả
kq1 = 1
kq2 = 1

# Luồng 1: nhân từ 1 đến n//2
def tinh_phan_1(n):
    global kq1
    for i in range(1, n//2 + 1):
        kq1 *= i

# Luồng 2: nhân từ n//2+1 đến n
def tinh_phan_2(n):
    global kq2
    for i in range(n//2 + 1, n + 1):
        kq2 *= i

n = 5

# Tạo 2 luồng
t1 = threading.Thread(target=tinh_phan_1, args=(n,))
t2 = threading.Thread(target=tinh_phan_2, args=(n,))

# Chạy luồng
t1.start()
t2.start()

# Chờ 2 luồng xong
t1.join()
t2.join()

# Kết quả cuối
giai_thua = kq1 * kq2
print(f"{n}! =", giai_thua)
```
