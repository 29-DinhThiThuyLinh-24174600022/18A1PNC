```python
import threading
import random

max_chung = -1
lock = threading.Lock()

# Hàm cho mỗi thread
def tim_max(thread_id, sub_list):
    global max_chung

    print(f"Thread {thread_id} :", end=" ")
    for x in sub_list:
        print(x, end=" ")
    print()

    max_con = max(sub_list)

    # Cập nhật max chung (dùng lock)
    lock.acquire()
    if max_con > max_chung:
        max_chung = max_con
    lock.release()


# Nhập số phần tử
n = int(input("Nhập số phần tử:\n"))
lst = [random.randint(0, 100) for _ in range(n)]
print("List:", lst)

# Nhập số thread
so_thread = int(input("Nhập vào số thread:\n"))

threads = []
kich_thuoc = n // so_thread

# Tạo thread
for i in range(so_thread):
    start = i * kich_thuoc
    end = (i + 1) * kich_thuoc if i != so_thread - 1 else n
    sub_list = lst[start:end]

    t = threading.Thread(target=tim_max, args=(i + 1, sub_list))
    threads.append(t)
    t.start()

# Chờ các thread hoàn thành
for t in threads:
    t.join()

print("Giá trị lớn nhất trong list =", max_chung)
```
