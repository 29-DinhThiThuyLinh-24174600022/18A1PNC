```python
import threading

# Luồng ghi số chẵn
def ghi_so_chan(nguong):
    print("Số chẵn:")
    for i in range(0, nguong + 1):
        if i % 2 == 0:
            print(i, end=" ")
    print()

# Luồng ghi số lẻ
def ghi_so_le(nguong):
    print("Số lẻ:")
    for i in range(0, nguong + 1):
        if i % 2 != 0:
            print(i, end=" ")
    print()

nguong = 20

# Tạo 2 luồng
t1 = threading.Thread(target=ghi_so_chan, args=(nguong,))
t2 = threading.Thread(target=ghi_so_le, args=(nguong,))

# Chạy luồng
t1.start()
t2.start()

# Chờ 2 luồng xong
t1.join()
t2.join()

print("Hoàn thành")
```
