```python
import threading
import time

def web(name, delay, so_lan):
    print(f"Starting {name}")
    for _ in range(so_lan):
        time.sleep(delay)
        print(f"{name}: Web {time.ctime()}")
    print(f"Exiting {name}")

print("Starting Main Thread")

# Tạo các luồng
t_google = threading.Thread(target=web, args=("Google", 1, 4))
t_yahoo = threading.Thread(target=web, args=("Yahoo", 2, 3))
t_facebook = threading.Thread(target=web, args=("Facebook", 3, 4))

# Chạy các luồng
t_google.start()
t_yahoo.start()
t_facebook.start()

# Chờ các luồng kết thúc
t_google.join()
t_yahoo.join()
t_facebook.join()

print("Exiting Main Thread")
```
