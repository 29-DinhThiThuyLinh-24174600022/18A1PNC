```python
import threading
import requests

def gui_request(url):
    print("Đang gửi request tới:", url)
    r = requests.get(url)
    print("Xong:", url, "| Status:", r.status_code)

urls = [
    "https://www.google.com",
    "https://www.python.org",
    "https://www.github.com"
]

threads = []

# Tạo và chạy các luồng
for url in urls:
    t = threading.Thread(target=gui_request, args=(url,))
    threads.append(t)
    t.start()

# Chờ tất cả luồng xong
for t in threads:
    t.join()

print("Hoàn thành tất cả request")
```
