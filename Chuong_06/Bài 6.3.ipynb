```python
import threading

# Luồng in số chẵn
def so_chan():
    print("Các số chẵn:")
    for i in range(30, 51):
        if i % 2 == 0:
            print(i, end=" ")
    print()

# Luồng in số lẻ
def so_le():
    print("Các số lẻ:")
    for i in range(30, 51):
        if i % 2 != 0:
            print(i, end=" ")
    print()

# Tạo 2 luồng
t1 = threading.Thread(target=so_chan)
t2 = threading.Thread(target=so_le)

# Chạy luồng
t1.start()
t2.start()

# Chờ 2 luồng chạy xong
t1.join()
t2.join()

print("Hoàn thành")
```
