```python
import threading
import time

# Hàm chạy cho mỗi luồng
def web(name, delay, times):
    print(f"Starting {name}")
    for _ in range(times):
        time.sleep(delay)
        print(f"{name}: Web {time.ctime()}")
    print(f"Exiting {name}")

print("Starting Main Thread")

# Tạo các luồng
t1 = threading.Thread(target=web, args=("Google", 1, 4))
t2 = threading.Thread(target=web, args=("Yahoo", 2, 4))
t3 = threading.Thread(target=web, args=("Facebook", 3, 4))

# Chạy luồng
t1.start()
t2.start()
t3.start()

# Chờ các luồng hoàn thành
t1.join()
t2.join()
t3.join()

print("Exiting Main Thread")
```
