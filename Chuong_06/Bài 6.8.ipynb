```python
import threading
import random

tong = 0
lock = threading.Lock()

# Hàm cho mỗi thread
def tinh_tong(thread_id, sub_list):
    global tong
    print(f"Thread\n{thread_id} :", end=" ")
    for x in sub_list:
        print(x, end="; ")
    print()

    sub_sum = sum(sub_list)

    # Cộng vào tổng chung (dùng lock)
    lock.acquire()
    tong += sub_sum
    lock.release()


# Nhập số phần tử
n = int(input("Nhập số phần tử:\n"))
lst = [random.randint(0, 10) for _ in range(n)]
print("List:", lst)

# Nhập số thread
so_thread = int(input("Nhập vào số thread:\n"))

threads = []
kich_thuoc = n // so_thread

# Tạo thread
for i in range(so_thread):
    start = i * kich_thuoc
    end = (i + 1) * kich_thuoc if i != so_thread - 1 else n
    sub_list = lst[start:end]

    t = threading.Thread(target=tinh_tong, args=(i + 1, sub_list))
    threads.append(t)
    t.start()

# Chờ các thread xong
for t in threads:
    t.join()

print("Tổng list =", tong)
```
