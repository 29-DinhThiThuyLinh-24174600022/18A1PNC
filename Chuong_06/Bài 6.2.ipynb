```python
import threading
import urllib.request

# Hàm tải file
def tai_file(url, ten_file):
    print("Bắt đầu tải:", ten_file)
    urllib.request.urlretrieve(url, ten_file)
    print("Tải xong:", ten_file)

# Danh sách các file cần tải
files = [
    ("https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf", "file1.pdf"),
    ("https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf", "file2.pdf"),
    ("https://www.w3.org/WAI/ER/tests/xhtml/testfiles/resources/pdf/dummy.pdf", "file3.pdf")
]

threads = []

# Tạo và chạy các luồng
for url, ten in files:
    t = threading.Thread(target=tai_file, args=(url, ten))
    threads.append(t)
    t.start()

# Chờ tất cả luồng hoàn thành
for t in threads:
    t.join()

print("Đã tải xong tất cả các file")
```
