```python
import numpy as np

baseball = []
with open("baseball_2D.txt", "r") as f:
    for line in f:
        values = line.strip().split()
        baseball.append([float(values[0]), float(values[1])])

np_baseball = np.array(baseball)

print("Kiểu dữ liệu của np_baseball:", np_baseball.dtype)
print("Kích thước của np_baseball:", np_baseball.shape)

print("Dòng thứ 50 trong np_baseball:", np_baseball[49])

np_weight = np_baseball[:, 1]
print("Mảng cân nặng:", np_weight)

height_124 = np_baseball[123, 0]
print("Chiều cao của vận động viên thứ 124:", height_124)

avg_height = np.mean(np_baseball[:, 0])
avg_weight = np.mean(np_baseball[:, 1])
print("Chiều cao trung bình:", avg_height)
print("Cân nặng trung bình:", avg_weight)

correlation = np.corrcoef(np_baseball[:, 0], np_baseball[:, 1])[0, 1]
print("Hệ số tương quan giữa chiều cao và cân nặng:", correlation)

if correlation > 0:
    print("→ Có tương quan thuận: cầu thủ cao hơn thường nặng hơn.")
elif correlation < 0:
    print("→ Có tương quan nghịch: cầu thủ cao hơn thường nhẹ hơn.")
else:
    print("→ Không có tương quan rõ ràng giữa chiều cao và cân nặng.")
```

## Bài 3.8

```python

heights_data = [
    180, 175, 190, 178, 185, 170, 182, 188, 176, 192,
    181, 174, 187, 179, 183, 186, 177, 189, 172, 184
]

positions_data = [
    "GK", "D", "M", "A", "GK", "D", "M", "A", "D", "A",
    "M", "GK", "A", "D", "M", "GK", "A", "D", "M", "A"
]

with open("heights.txt", "w") as f:
    for h in heights_data:
        f.write(str(h) + "\n")

with open("positions.txt", "w") as f:
    for p in positions_data:
        f.write(p + "\n")

print(" Đã tạo xong 2 tập tin heights.txt và positions.txt")
```

```python
import numpy as np

with open("heights.txt") as f:
    heights = [float(line.strip()) for line in f]

with open("positions.txt") as f:
    positions = [line.strip() for line in f]

np_positions = np.array(positions)
np_heights = np.array(heights)

print("Kiểu dữ liệu np_positions:", np_positions.dtype)
print("Kiểu dữ liệu np_heights:", np_heights.dtype)

mean_GK = np.mean(np_heights[np_positions == "GK"])
print("Chiều cao trung bình của GK:", round(mean_GK, 2))

mean_not_GK = np.mean(np_heights[np_positions != "GK"])
print("Chiều cao trung bình của các vị trí khác GK:", round(mean_not_GK, 2))

players = np.array(list(zip(np_positions, np_heights)),
                   dtype=[('position', 'U5'), ('height', 'float')])

players_sorted = np.sort(players, order='height')
print("Chiều cao thấp nhất:", players_sorted[0])
print("Chiều cao cao nhất:", players_sorted[-1])
```

```python
import pandas as pd

data = {
    'country': ['USA', 'France', 'Germany', 'Italy', 'Spain', 'UK', 'China', 'India', 'Brazil', 'Australia'],
    'continent': ['North America', 'Europe', 'Europe', 'Europe', 'Europe', 'Europe', 'Asia', 'Asia', 'South America', 'Oceania'],
    'beer_servings': [229, 127, 346, 85, 193, 245, 40, 18, 245, 261],
    'spirit_servings': [87, 99, 175, 123, 101, 89, 40, 15, 90, 72],
    'wine_servings': [50, 312, 175, 218, 250, 101, 35, 12, 20, 85],
    'total_litres_of_pure_alcohol': [9.8, 12.3, 11.5, 12.9, 12.1, 11.7, 5.1, 4.3, 7.0, 10.0]
}

df = pd.DataFrame(data)

df.to_csv('drinks.csv', index=False)

print("File drinks.csv đã được tạo.")
```

```python

drink = pd.read_csv('drinks.csv', index_col=0)


print("Kiểu dữ liệu:", type(drink))
print("Kích thước (shape):", drink.shape)

print("Các cột:", drink.columns.tolist())

print("\n5 dòng đầu (head):\n", drink.head())
print("\n5 dòng cuối (tail):\n", drink.tail())

mean_beer = drink.groupby('continent')['beer_servings'].mean()
print("\nSố lượng bia tiêu thụ trung bình theo châu lục:\n", mean_beer)

wine_stats = drink.groupby('continent')['wine_servings'].describe()
print("\nThông tin thống kê rượu vang theo châu lục:\n", wine_stats)

mean_drinks = drink.groupby('continent')[['beer_servings','wine_servings','spirit_servings']].mean()
print("\nTrung bình bia, rượu vang, rượu mạnh theo châu lục:\n", mean_drinks)

median_drinks = drink.groupby('continent')[['beer_servings','wine_servings','spirit_servings']].median()
print("\nTrung vị bia, rượu vang, rượu mạnh theo châu lục:\n", median_drinks)

spirit_stats = drink.groupby('continent')['spirit_servings'].agg(['mean','max','min'])
print("\nSpirit servings - mean, max, min theo châu lục:\n", spirit_stats)
sorted_beer = drink.sort_values(by='beer_servings', ascending=True)

print("\n5 quốc gia tiêu thụ bia nhiều nhất:\n", sorted_beer.tail(5))
print("\n5 quốc gia tiêu thụ bia ít nhất:\n", sorted_beer.head(5))
```

```python
import pandas as pd

# Tạo stocks1.csv
stocks1_data = {
    'date': ['2025-01-01','2025-01-02','2025-01-03','2025-01-04','2025-01-05'],
    'symbol': ['AAPL','AAPL','GOOG','GOOG','MSFT'],
    'open': [150, 152, 1200, 1195, 210],
    'high': [151, None, 1205, 1200, 215],  # có giá trị None để test Null
    'low': [149, 151, None, 1185, 208],    # có giá trị None để test Null
    'close': [150.5, 151.5, 1202, 1198, 212],
    'volume': [10000, 12000, 5000, 6000, 8000]
}
pd.DataFrame(stocks1_data).to_csv('stocks1.csv', index=False)

# Tạo stocks2.csv
stocks2_data = {
    'date': ['2025-01-01','2025-01-02','2025-01-03','2025-01-04','2025-01-05'],
    'symbol': ['FB','FB','TSLA','TSLA','AMZN'],
    'open': [270, 272, 650, 655, 3100],
    'high': [275, 274, 655, 660, 3120],
    'low': [268, 269, 645, 650, 3090],
    'close': [272, 273, 652, 658, 3110],
    'volume': [7000, 8000, 4000, 4500, 9000]
}
pd.DataFrame(stocks2_data).to_csv('stocks2.csv', index=False)

# Tạo companies.csv
companies_data = {
    'name': ['AAPL','GOOG','MSFT','FB','TSLA','AMZN'],
    'employees': [150000, 135000, 163000, 58000, 70000, 133000],
    'headquarters_city': ['Cupertino','Mountain View','Redmond','Menlo Park','Palo Alto','Seattle'],
    'headquarters_state': ['CA','CA','WA','CA','CA','WA']
}
pd.DataFrame(companies_data).to_csv('companies.csv', index=False)

print("Đã tạo xong các file CSV: stocks1.csv, stocks2.csv, companies.csv")
```

```python
# 1a) 
stocks1 = pd.read_csv('stocks1.csv')
print("stocks1 - 5 dòng đầu:\n", stocks1.head())
print("stocks1 - 5 dòng cuối:\n", stocks1.tail())
print("stocks1 dtypes:\n", stocks1.dtypes)
print("stocks1 info:")
stocks1.info()

# 1b) 
stocks2 = pd.read_csv('stocks2.csv')
print("\nstocks2 - 5 dòng đầu:\n", stocks2.head())
print("stocks2 - 5 dòng cuối:\n", stocks2.tail())
print("stocks2 dtypes:\n", stocks2.dtypes)
print("stocks2 info:")
stocks2.info()

# 1c)
companies = pd.read_csv('companies.csv')
print("\ncompanies data:\n", companies)
print("companies dtypes:\n", companies.dtypes)
print("companies info:")
companies.info()

#
print("\nNull values in stocks1:\n", stocks1.isnull().sum())
stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))
print("\nSau khi thay thế Null:\n", stocks1)

stocks = pd.concat([stocks1, stocks2], ignore_index=True)
print("\n15 dòng cuối của stocks:\n", stocks.tail(15))

stocks_companies = pd.merge(stocks, companies, left_on='symbol', right_on='name', how='left')
print("\n5 dòng đầu stocks_companies:\n", stocks_companies.head())

mean_values = stocks_companies.groupby('symbol')[['open','high','low','close','volume']].mean()
print("\nGiá trung bình và volume trung bình mỗi công ty:\n", mean_values)

close_stats = stocks_companies.groupby('symbol')['close'].agg(['mean','max','min'])
print("\nClose price - mean, max, min mỗi công ty:\n", close_stats)

stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print("\nKiểu dữ liệu của parsed_time:", stocks_companies['parsed_time'].dtype)
print("5 dòng đầu:\n", stocks_companies[['date','parsed_time']].head())

stocks_companies['result'] = stocks_companies.apply(lambda row: 'up' if row['close'] > row['open'] else 'down', axis=1)
print("\n5 dòng đầu với cột result:\n", stocks_companies[['symbol','open','close','result']].head())
```

```python
import os
import urllib.request
import zipfile

# 1. Xác định URL và nơi lưu trữ
url = "https://grouplens.org/datasets/movielens/latest/"  # đường dẫn tải chính thức  [oai_citation:0‡GroupLens](https://grouplens.org/datasets/movielens/latest/?utm_source=chatgpt.com)
download_path = "ml-latest-small.zip"
extract_dir = "ml-latest-small"

# 2. Nếu file chưa tồn tại thì tải về
if not os.path.exists(download_path):
    print("Đang tải dữ liệu từ:", url)
    urllib.request.urlretrieve(url, download_path)
    print("Tải xong:", download_path)
else:
    print("File đã tồn tại:", download_path)

# 3. Giải nén nếu thư mục chứa dữ liệu chưa được tạo
if not os.path.exists(extract_dir):
    print("Đang giải nén vào thư mục:", extract_dir)
    with zipfile.ZipFile(download_path, 'r') as zip_ref:
        zip_ref.extractall(extract_dir)
    print("Giải nén xong.")
else:
    print("Thư mục đã tồn tại:", extract_dir)

# 4. Xác nhận các file chính
print("Các file trong thư mục:", os.listdir(extract_dir))
```
