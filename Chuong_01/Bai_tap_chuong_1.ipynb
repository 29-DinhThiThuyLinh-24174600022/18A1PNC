```python
class HinhChuNhat:
    def __init__(self):
        self.dai = 0
        self.rong = 0

    def nhap(self):
        self.dai = float(input("Nhập chiều dài: "))
        self.rong = float(input("Nhập chiều rộng: "))

    def tinh_chu_vi(self):
        return 2 * (self.dai + self.rong)

    def tinh_dien_tich(self):
        return self.dai * self.rong

    def hien_thi(self):
        print("\n--- Thông tin hình chữ nhật ---")
        print("Chiều dài:", self.dai)
        print("Chiều rộng:", self.rong)
        print("Chu vi:", self.tinh_chu_vi())
        print("Diện tích:", self.tinh_dien_tich())


# Chương trình chính
hcn = HinhChuNhat()
hcn.nhap()
hcn.hien_thi()
```

```python
class ThiSinh:
    def __init__(self):
        self.hoten = ""
        self.toan = 0
        self.ly = 0
        self.hoa = 0

    def nhap(self):
        self.hoten = input("Nhập họ tên thí sinh: ")
        self.toan = float(input("Nhập điểm Toán: "))
        self.ly = float(input("Nhập điểm Lý: "))
        self.hoa = float(input("Nhập điểm Hóa: "))

    def tinh_tong(self):
        return self.toan + self.ly + self.hoa

    def hien_thi(self):
        print(f"Họ tên: {self.hoten}")
        print(f"Toán: {self.toan}, Lý: {self.ly}, Hóa: {self.hoa}")
        print(f"Tổng điểm: {self.tinh_tong()}")
        print("--------------------------")

ds = []
n = int(input("Nhập số lượng thí sinh: "))

for i in range(n):
    print(f"\n-- Thí sinh thứ {i+1} --")
    ts = ThiSinh()
    ts.nhap()
    ds.append(ts)

diem_chuan = float(input("\nNhập điểm chuẩn: "))

print("\n=== DANH SÁCH THÍ SINH TRÚNG TUYỂN ===")
for ts in ds:
    if ts.tinh_tong() >= diem_chuan:
        ts.hien_thi()
```

```python
class PhanSo:
    def __init__(self):
        self.tu = 0
        self.mau = 1

    def kiem_tra_hop_le(self):
        if self.mau == 0:
            return False
        return True

    def nhap(self):
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        # kiểm tra mẫu hợp lệ
        while self.mau == 0:
            print("Mẫu số phải khác 0, vui lòng nhập lại!")
            self.mau = int(input("Nhập mẫu số: "))

    def hien_thi(self):
        if not self.kiem_tra_hop_le():
            print("Phân số không hợp lệ!")
        else:
            print(f"Phân số là: {self.tu}/{self.mau}")

# === Chương trình chính ===
ps = PhanSo()
ps.nhap()
ps.hien_thi()
```

```python
class Stack:
    def __init__(self, max_size):
        self.max_size = max_size
        self.stack = []

    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) == self.max_size

    def push(self, value):
        if self.isFull():
            print(" Ngăn xếp đã đầy, không thể thêm!")
        else:
            self.stack.append(value)
            print(f" Đã thêm {value} vào ngăn xếp.")

    def pop(self):
        if self.isEmpty():
            print(" Ngăn xếp rỗng, không thể lấy phần tử!")
        else:
            value = self.stack.pop()
            print(f"Đã lấy ra phần tử: {value}")

    def hien_thi(self):
        if self.isEmpty():
            print(" Ngăn xếp hiện đang rỗng.")
        else:
            print(" Ngăn xếp hiện tại:", self.stack)


# === Chương trình chính ===
stack = Stack(int(input("Nhập kích thước tối đa của ngăn xếp: ")))

while True:
    print("\n===== MENU NGĂN XẾP =====")
    print("1. Thêm phần tử (push)")
    print("2. Lấy phần tử (pop)")
    print("3. Hiển thị ngăn xếp")
    print("4. Thoát")
    chon = input("Nhập lựa chọn của bạn (1-4): ")

    if chon == "1":
        x = float(input("Nhập giá trị cần thêm: "))
        stack.push(x)
    elif chon == "2":
        stack.pop()
    elif chon == "3":
        stack.hien_thi()
    elif chon == "4":
        print(" Kết thúc chương trình.")
        break
    else:
        print(" Lựa chọn không hợp lệ, vui lòng nhập lại.")
```

```python
class Stack:
    def __init__(self, size):
        self.size = size
        self.stack = []

    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) == self.size

    def push(self, value):
        if self.isFull():
            print(" Ngăn xếp đã đầy, không thể thêm phần tử.")
        else:
            self.stack.append(value)
            print(f" Đã thêm {value} vào ngăn xếp.")

    def pop(self):
        if self.isEmpty():
            print(" Ngăn xếp rỗng, không thể lấy phần tử.")
        else:
            value = self.stack.pop()
            print(f" Đã lấy ra phần tử: {value}")
            return value

    def count(self):
        return len(self.stack)

    # ✅ Hàm mới: Hiển thị toàn bộ nội dung ngăn xếp
    def hien_thi(self):
        if self.isEmpty():
            print(" Ngăn xếp rỗng, không có gì để hiển thị.")
        else:
            print(" Nội dung ngăn xếp (từ dưới lên trên):")
            for i in range(len(self.stack)):
                print(f"  {self.stack[i]}")
            print(" Phần tử trên cùng là:", self.stack[-1])


# === Chương trình chính ===
n = int(input("Nhập kích thước ngăn xếp: "))
st = Stack(n)

so_luong = int(input("Nhập số phần tử muốn thêm vào ngăn xếp: "))
for i in range(so_luong):
    x = float(input(f"Nhập phần tử thứ {i+1}: "))
    st.push(x)

print("\nSố phần tử hiện có:", st.count())
st.hien_thi()
```

```python
class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0)

    # Trả về số ngày trong tháng
    def days_in_month(self):
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif self.month in [4, 6, 9, 11]:
            return 30
        elif self.month == 2:
            if self.is_leap_year():
                return 29
            else:
                return 28

    # Phương thức hiển thị ngày tháng năm
    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")

    # Phương thức tính ngày kế tiếp
    def next(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1

d = int(input("Nhập ngày: "))
m = int(input("Nhập tháng: "))
y = int(input("Nhập năm: "))

today = Date(d, m, y)

print("\nNgày hiện tại là:")
today.display()

today.next()
print("Ngày kế tiếp là:")
today.display()
```

```python
class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year

    def is_leap_year(self):
        return (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0)

    def days_in_month(self):
        if self.month in [1, 3, 5, 7, 8, 10, 12]:
            return 31
        elif self.month in [4, 6, 9, 11]:
            return 30
        elif self.month == 2:
            return 29 if self.is_leap_year() else 28

    def display(self):
        return f"{self.day:02d}/{self.month:02d}/{self.year}"

    def next(self):
        self.day += 1
        if self.day > self.days_in_month():
            self.day = 1
            self.month += 1
            if self.month > 12:
                self.month = 1
                self.year += 1


# --- Lớp Employee ---
class Employee:
    def __init__(self, ho_ten, ngay_sinh, ngay_vao_cty):
        self.ho_ten = ho_ten
        self.ngay_sinh = ngay_sinh
        self.ngay_vao_cty = ngay_vao_cty

    def display(self):
        print(f"Họ tên: {self.ho_ten}")
        print(f"Ngày sinh: {self.ngay_sinh.display()}")
        print(f"Ngày vào công ty: {self.ngay_vao_cty.display()}")
        print("-" * 30)


ds_nv = []  

n = int(input("Nhập số lượng nhân viên: "))

for i in range(n):
    print(f"\nNhập thông tin nhân viên thứ {i+1}:")
    ho_ten = input("Họ tên: ")

    # Nhập ngày sinh
    d1 = int(input("Ngày sinh: "))
    m1 = int(input("Tháng sinh: "))
    y1 = int(input("Năm sinh: "))
    ngay_sinh = Date(d1, m1, y1)

    # Nhập ngày vào công ty
    d2 = int(input("Ngày vào công ty: "))
    m2 = int(input("Tháng vào công ty: "))
    y2 = int(input("Năm vào công ty: "))
    ngay_vao = Date(d2, m2, y2)

    nv = Employee(ho_ten, ngay_sinh, ngay_vao)
    ds_nv.append(nv)

# --- Hiển thị danh sách nhân viên ---
print("\n===== DANH SÁCH NHÂN VIÊN =====")
for nv in ds_nv:
    nv.display()
```

```python
import math

# Lớp da giac
class DaGiac:
    def __init__(self, so_canh):
        self.so_canh = so_canh
        self.ds_canh = []

    def nhap_canh(self):
        print(f"Nhập độ dài {self.so_canh} cạnh:")
        for i in range(self.so_canh):
            canh = float(input(f"Cạnh {i+1}: "))
            self.ds_canh.append(canh)

    def tinh_chu_vi(self):
        return sum(self.ds_canh)

# Lớp hinh binh hanh ke thua tu da giac
class HinhBinhHanh(DaGiac):
    def __init__(self, day_lon, day_nho, chieu_cao):
        super().__init__(4)
        self.day_lon = day_lon
        self.day_nho = day_nho
        self.chieu_cao = chieu_cao

    def tinh_chu_vi(self):
        return 2 * (self.day_lon + self.day_nho)

    def tinh_dien_tich(self):
        return self.day_lon * self.chieu_cao

# Lớp hinh chu nhat ke thua tu hinh binh hanh
class HinhChuNhat(HinhBinhHanh):
    def __init__(self, chieu_dai, chieu_rong):
        super().__init__(chieu_dai, chieu_rong, chieu_rong)
        self.chieu_dai = chieu_dai
        self.chieu_rong = chieu_rong

    def tinh_dien_tich(self):
        return self.chieu_dai * self.chieu_rong

# Lớp hinh vuong ke thua tu hinh chu nhat
class HinhVuong(HinhChuNhat):
    def __init__(self, canh):
        super().__init__(canh, canh)
        self.canh = canh

    def tinh_dien_tich(self):
        return self.canh ** 2

# --- Chuong trinh chinh ---
print("=== TÍNH CHU VI & DIỆN TÍCH CÁC HÌNH ===")

# Hinh binh hanh
print("\nHình bình hành:")
a = float(input("Nhập đáy lớn: "))
b = float(input("Nhập cạnh bên: "))
h = float(input("Nhập chiều cao: "))
hbh = HinhBinhHanh(a, b, h)
print("Chu vi:", hbh.tinh_chu_vi())
print("Diện tích:", hbh.tinh_dien_tich())

# Hinh chu nhat
print("\nHình chữ nhật:")
d = float(input("Nhập chiều dài: "))
r = float(input("Nhập chiều rộng: "))
hcn = HinhChuNhat(d, r)
print("Chu vi:", hcn.tinh_chu_vi())
print("Diện tích:", hcn.tinh_dien_tich())

# Hinh vuong
print("\nHình vuông:")
c = float(input("Nhập cạnh: "))
hv = HinhVuong(c)
print("Chu vi:", hv.tinh_chu_vi())
print("Diện tích:", hv.tinh_dien_tich())
```

```python
import math

class Diem:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def hien_thi(self):
        print(f"Tọa độ điểm: ({self.x}, {self.y})")

# Lop elip ke thua tu diem
class Elip(Diem):
    def __init__(self, x, y, ban_truc_lon, ban_truc_nho):
        super().__init__(x, y)
        self.ban_truc_lon = ban_truc_lon
        self.ban_truc_nho = ban_truc_nho

    def tinh_dien_tich(self):
        return math.pi * self.ban_truc_lon * self.ban_truc_nho

# Lop duong tron ke thua tu elip
class DuongTron(Elip):
    def __init__(self, x, y, ban_kinh):
        super().__init__(x, y, ban_kinh, ban_kinh)
        self.ban_kinh = ban_kinh

    def tinh_dien_tich(self):
        return math.pi * self.ban_kinh ** 2

# --- Chuong trinh chinh ---
print("=== TÍNH DIỆN TÍCH ELIP & ĐƯỜNG TRÒN ===")

# Nhap elip
print("\nHình elip:")
x = float(input("Nhập hoành độ tâm: "))
y = float(input("Nhập tung độ tâm: "))
a = float(input("Nhập bán trục lớn a: "))
b = float(input("Nhập bán trục nhỏ b: "))
elip = Elip(x, y, a, b)
elip.hien_thi()
print("Diện tích elip =", round(elip.tinh_dien_tich(), 2))

# Nhap duong tron
print("\nHình tròn:")
x = float(input("Nhập hoành độ tâm: "))
y = float(input("Nhập tung độ tâm: "))
r = float(input("Nhập bán kính r: "))
tron = DuongTron(x, y, r)
tron.hien_thi()
print("Diện tích đường tròn =", round(tron.tinh_dien_tich(), 2))
```

```python
import math

# Lop tam giac
class TamGiac:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def kiem_tra(self):
        return (self.a + self.b > self.c) and (self.a + self.c > self.b) and (self.b + self.c > self.a)

    def tinh_chu_vi(self):
        if not self.kiem_tra():
            return None
        return self.a + self.b + self.c

    def tinh_dien_tich(self):
        if not self.kiem_tra():
            return None
        p = self.tinh_chu_vi() / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

# Lop tam giac vuong ke thua tu TamGiac
class TamGiacVuong(TamGiac):
    def la_tam_giac_vuong(self):
        canh = sorted([self.a, self.b, self.c])
        return math.isclose(canh[2]**2, canh[0]**2 + canh[1]**2)

# Lop tam giac can ke thua tu TamGiac
class TamGiacCan(TamGiac):
    def la_tam_giac_can(self):
        return self.a == self.b or self.a == self.c or self.b == self.c

# Lop tam giac deu ke thua tu TamGiacCan
class TamGiacDeu(TamGiacCan):
    def la_tam_giac_deu(self):
        return self.a == self.b == self.c

    def tinh_dien_tich(self):
        if not self.la_tam_giac_deu():
            return None
        # Công thức riêng cho tam giác đều
        return (math.sqrt(3) / 4) * self.a**2

# --- Chuong trinh chinh ---
print("=== TÍNH CHU VI & DIỆN TÍCH TAM GIÁC ===")

a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

tg = TamGiac(a, b, c)
if not tg.kiem_tra():
    print("Ba cạnh không tạo thành tam giác!")
else:
    print("Chu vi tam giác:", tg.tinh_chu_vi())
    print("Diện tích tam giác:", round(tg.tinh_dien_tich(), 2))

    tg_vuong = TamGiacVuong(a, b, c)
    tg_can = TamGiacCan(a, b, c)
    tg_deu = TamGiacDeu(a, b, c)

    if tg_vuong.la_tam_giac_vuong():
        print("Đây là tam giác vuông.")
    elif tg_deu.la_tam_giac_deu():
        print("Đây là tam giác đều.")
        print("Diện tích tam giác đều:", round(tg_deu.tinh_dien_tich(), 2))
    elif tg_can.la_tam_giac_can():
        print("Đây là tam giác cân.")
    else:
        print("Đây là tam giác thường.")
```
